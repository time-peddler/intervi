```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# sns.set_context("paper")

np.random.seed(2020)
```

```python
from datetime import datetime
```

```python
df1 = pd.read_excel('./面试题目2资料.xlsx',0)
```

```python
df1.columns
```

## 1.评级高的3种产品

```python
rank = df1.sort_values('产品评级', ascending=False).head(3)
```

## 2.收藏最多的8种产品

```python
fav = df1.sort_values('产品收藏记录', ascending=False).head(8)
```

## 3.刊登类目最多的5种商品

```python
publish = df1.sort_values('刊登类目', ascending=False).head(5)
```

## 4.销售最多的10种商品

```python
sale = df1.sort_values('销售记录', ascending=False).head(10)
```

## 5.加权优秀商品（评级20%,收藏30%，销售40%，刊登10%）

```python
df1['weight'] = df1['产品评级']*0.2 + df1['产品收藏记录']*0.3 + df1['刊登类目']/10000*0.2 + df1['销售记录']*0.4
```

```python
weight = df1.sort_values('weight', ascending=False).head(10)
```

## 6.价格最高和最低的各3种产品

```python
price = pd.concat([df1.nlargest(3, '价格'),df1.nsmallest(3,'价格')])
```

# ！！ 最终进到系统中销售的产品

```python
result = pd.concat([sale,publish,fav,rank,weight,price]).drop_duplicates()
```
